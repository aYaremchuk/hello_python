```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import pandas as pd
```

```python
data = pd.read_csv('countries.csv')
```

```python
data.nlargest(10, 'Population')
```

```python
data.nlargest(10, 'Population').plot('Country', 'Population', kind='bar')
```

```python
data.Region.value_counts()
```

```python
data.Region.value_counts().plot(kind='barh')
```

```python
data.sort_values(by=['Area (sq. mi.)'], ascending=False).head(10)
```

```python
data.sort_values(by=['Area (sq. mi.)'], ascending=False).head(10).plot('Country','Area (sq. mi.)', kind='barh')
```
